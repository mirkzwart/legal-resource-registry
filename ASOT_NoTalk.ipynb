```python
#authorization
#tk = spotipy.util.prompt_for_user_token(username, scope, client_id, client_secret, redirect_uri)
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    scope=scope,
    username=username,
    redirect_uri=redirect_uri,
    client_id=client_id,
    client_secret=client_secret,
    open_browser=False))
```

```python
#function to call playlists
def call_playlist(creator, playlist_id):
    playlist_features_list = ['artist','album','track_name','track_id','duration','uri']
    playlist_df = pd.DataFrame(columns = playlist_features_list)
    
    playlist = sp.user_playlist_tracks(creator, playlist_id, offset=0)['items']
    for track in playlist:
        playlist_features = {}
        playlist_features["artist"] = track["track"]["album"]["artists"][0]["name"]
        playlist_features["album"] = track["track"]["album"]["name"]
        playlist_features["track_name"] = track["track"]["name"]
        playlist_features["track_id"] = track["track"]["id"]
        playlist_features["uri"] = track["track"]["uri"]
        playlist_features["duration"] = track["track"]["duration_ms"]
    
        track_df = pd.DataFrame(playlist_features, index = [0])
        playlist_df = pd.concat([playlist_df, track_df], ignore_index = True)
        
    return playlist_df
```

```python
#call 100 items of current full playlist and drop empty rows
asot_all = call_playlist('spotify', asot_id)
asot_all.dropna(inplace = True)

#create episode column and make it the index
start, stop, step = 0,9,1
asot_all['episode'] = asot_all['album'].str.slice(start, stop, step)
asot_all_by_episode = asot_all.set_index('episode')

#drop all songs under 55 seconds
asot_all_by_episode_long = asot_all_by_episode[asot_all_by_episode.duration >= 55000]

asot_all_by_episode_long
```

```python
#call 100 items of the no talk playlist
asot_no_talk_items = call_playlist('spotify', asot_no_talk)
asot_no_talk_items.dropna(inplace = True)

#create episode column and make it the index
start, stop, step = 0,9,1
asot_no_talk_items['episode'] = asot_no_talk_items['album'].str.slice(start, stop, step)
asot_no_talk_items_by_episode = asot_no_talk_items.set_index('episode')
asot_no_talk_items_by_episode
```

```python
#compare databases
df_all = asot_all_by_episode_long.merge(asot_no_talk_items_by_episode.drop_duplicates(), on=['episode','album'], how='left', indicator=True)
#keep only the new songs
df_new = df_all[df_all._merge == 'left_only']
```

```python
#create list of uris and push to spotify
df_new_list = df_new['uri_x'].tolist()

sp.playlist_add_items(asot_no_talk, df_new_list, position=0)
```
